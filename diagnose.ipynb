```python
df = pd.DataFrame(data)
```

```python
df = df.drop(columns=['samples'])
```

```python
df = shuffle(df)
df.head(10)
```

```python
df = df.dropna(axis=1)
df = df.dropna()
```

```python
labels_count = df['type'].value_counts()
print(labels_count)
```

```python
label_encoder = LabelEncoder()
```

```python
df.iloc[:, 0] = label_encoder.fit_transform(df.iloc[:, 0].values)
df.head(10)
```

```python
df.iloc[:, 1:] = np.log2(df.iloc[:, 1:].values)
```

One-way ANOVA test

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
df.head()
```

```python
features = df.iloc[:, 1:].columns.values
features
```

```python
y = y.astype('int')
(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.33, random_state=42)
X_test
```

```python
sc=StandardScaler()
X_train = pd.DataFrame(sc.fit_transform(X_train.values), columns=X_train.columns)
X_test = pd.DataFrame(sc.fit_transform(X_test.values), columns=X_test.columns)
```

```python
fs = SelectKBest(score_func=f_classif, k='all').fit(X_train, y_train)
features_score = pd.DataFrame(fs.scores_)
features_pvalue = pd.DataFrame(fs.pvalues_)
features = pd.DataFrame(X.columns)
features_score = pd.concat([features, features_score, features_pvalue], axis=1)
features_score.columns = ["Input_Features","F_score","P_Value"]
#features_score = features_score.iloc[:200, :]
features_score = features_score.sort_values('F_score').iloc[22200:, :]
features_score
```

```python
features_score.to_csv('top_features.csv')
```

```python
y_train
```

```python
important_features_extract = features_score.head(10).to_csv('features_extract.csv')
important_features_extract
```

```python
important_features = features_score.iloc[:, 0].to_list()
dt = df
important_features
```

```python
features_to_drop = [col for col in X_train.columns if col not in important_features]
X_train = X_train.drop(columns=features_to_drop)
X_test = X_test.drop(columns=features_to_drop)
X_test = sc.fit_transform(X_test.values)
```
